```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
from pyhht.emd import EMD
from pyhht.visualization import plot_imfs
import scipy.fftpack
%matplotlib inline
style.use('seaborn-white')
```

```python
x = np.linspace(-5, 5, 100)
t = range(-50,50)
function_sum = np.sin(2 * x) + 2*np.cos((5*x)**3)
 
decomposer = EMD(function_sum)
imfs = decomposer.decompose()
plot_imfs(function_sum, imfs, t)
```

```python
signal = np.loadtxt('eeg_ex.txt') 
decomposer = EMD(signal,)
imfs = decomposer.decompose()
plot_imfs(signal, imfs)
```

```python
def plotSignal(signal,start,end,title):
    plt.plot(range(start,end), signal[start:end])
    plt.title(title)
    plt.show()
    
def plotFFT(signal, freq, title):
    N = len(signal)
    T = 1.0 / freq
    yf = scipy.fftpack.fft(signal)
    xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
    fig, ax = plt.subplots()
    ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
    plt.title(title)
    plt.show()

def noiseMode(noise_modes):
    noise = 0
    for i in noise_modes:
        print("Minus")
        plotSignal(imfs[i],start,end,'Mode ' + str(i+1))
        plotFFT(imfs[i],250.0,'FFT of mode ' + str(i+1))
        noise += imfs[i]
    return noise
   
start = 4000
end = 6000

noise_modes = [0,5,6,7,8,9]

plotSignal(signal,start,end,'Raw signal')
plotFFT(signal,250.0,'FFT of signal')
filtered_EEG = signal - noiseMode(noise_modes)
print("Equals")
plotSignal(filtered_EEG,start,end,'Filtered signal')
```
